```python
def intermidiateNeighbor(Pattern): 
  mer_set = [Pattern]
  words = ['A','T','C','G']
  for i in range(len(Pattern)):
    char = Pattern[i]
    for symbol in words:
      if symbol != char:
        # should replace and add to mers 
        new_mer = Pattern[0:i] + symbol + Pattern[i+1:]
        mer_set.append(new_mer)

  return mer_set


def interativeNeighbor(Pattern, d):
  mer_set = set([Pattern])
  for i in range(d):
    mer_set_copy = mer_set.copy()
    for mer in mer_set_copy:
      # generate its neighbors
      all_list = set(intermidiateNeighbor(mer))
      for a in all_list:
        mer_set.add(a)
  return mer_set
```

```python
for a in interativeNeighbor('TAATTAATATC',3):
  print(a)
```
